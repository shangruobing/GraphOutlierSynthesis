```python
import matplotlib.pyplot as plt
import seaborn as sns
import csv
import numpy as np
import warnings
```

```python

warnings.filterwarnings('ignore')

f = csv.reader(open('./GNNSafe/results/vis_scores/arxiv-gnnsafe_no_prop_no_reg.csv', 'r'))
l = [i for i in f]
pos_num, neg_num = l[0][0].split(' ')
pos_num, neg_num = int(pos_num), int(neg_num)
score = -np.array(l[1:], dtype=np.float).reshape(-1)
print(score)
pos_score1, neg_score1 = score[:pos_num], score[pos_num:]

f = csv.reader(open('./GNNSafe/results/vis_scores/arxiv-gnnsafe_use_prop_no_reg.csv', 'r'))
l = [i for i in f]
pos_num, neg_num = l[0][0].split(' ')
pos_num, neg_num = int(pos_num), int(neg_num)
score = -np.array(l[1:], dtype=np.float).reshape(-1)
print(score)
pos_score2, neg_score2 = score[:pos_num], score[pos_num:]

f = csv.reader(open('./GNNSafe/results/vis_scores/arxiv-gnnsafe_use_prop_use_reg.csv', 'r'))
l = [i for i in f]
pos_num, neg_num = l[0][0].split(' ')
pos_num, neg_num = int(pos_num), int(neg_num)
score = -np.array(l[1:], dtype=np.float).reshape(-1)
print(score)
pos_score3, neg_score3 = score[:pos_num], score[pos_num:]

sns.set(style='white')
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 3))

# plot vi frequency distribution
sns.distplot(pos_score1, ax=ax[0], hist=False, kde_kws={'fill': True}, kde=True, color='green', label='in-distribution')
ax[0].vlines(np.median(pos_score1), 0, 0.8, color='green', linestyles='dashed')
sns.distplot(neg_score1, ax=ax[0], hist=False, kde_kws={'fill': True}, kde=True, color='red', label='out-of-distribution')
ax[0].vlines(np.median(neg_score1), 0, 0.8, color='red', linestyles='dashed')
ax[0].set_ylim(0,0.8)
ax[0].set_xlim(-15, 0)
ax[0].set_title('GNNSafe w/o energy propagation', size=15)
ax[0].set_xlabel('Energy score', size=15)
ax[0].set_ylabel('Frequency', size=15)
ax[0].legend(loc='upper left', fontsize=15)
# fig.savefig('./figs/arxiv-score-energy.pdf', bbox_inches='tight')

# plot base frequency distribution
sns.distplot(pos_score2, ax=ax[1], hist=False, kde_kws={'fill': True}, color='green', label='in-distribution')
ax[1].vlines(np.median(pos_score2), 0, 1.6, color='green', linestyles='dashed')
sns.distplot(neg_score2, ax=ax[1], hist=False, kde_kws={'fill': True}, color='red', label='out-of-distribution')
ax[1].vlines(np.median(neg_score2), 0, 1.6, color='red', linestyles='dashed')
ax[1].set_ylim(0,1.6)
ax[1].set_xlim(-10, 0)
ax[1].set_title('GNNSafe', size=15)
ax[1].set_xlabel('Energy score', size=15)
ax[1].set_ylabel('Frequency', size=15)
ax[1].legend(loc='upper left', fontsize=15)
# fig.savefig('./figs/arxiv-score-energyprop.pdf', bbox_inches='tight')

# plot 123 frequency distribution
sns.distplot(pos_score3, ax=ax[2], hist=False, kde_kws={'fill': True}, color='green', label='in-distribution')
ax[2].vlines(np.median(pos_score3), 0, 0.8, color='green', linestyles='dashed')
sns.distplot(neg_score3, ax=ax[2], hist=False, kde_kws={'fill': True}, color='red', label='out-of-distribution')
ax[2].vlines(np.median(neg_score3), 0, 0.8, color='red', linestyles='dashed')
ax[2].set_ylim(0,0.8)
ax[2].set_xlim(-10, 0)
ax[2].set_title('GNNSafe++', size=15)
ax[2].set_xlabel('Energy score', size=15)
ax[2].set_ylabel('Frequency', size=15)
ax[2].legend(loc='upper left', fontsize=15)
# fig.savefig('./figs/arxiv-score-energypropreg.pdf', bbox_inches='tight')

fig.tight_layout()
# fig.savefig('./figs/arxiv-score.pdf', bbox_inches='tight')
```

```python
# different backbones

import matplotlib.pyplot as plt
import seaborn as sns
import csv
import numpy as np

auroc1, aupr1, fpr1 = [], [], []
auroc2, aupr2, fpr2 = [], [], []
auroc3, aupr3, fpr3 = [], [], []

f = csv.reader(open('./GNNSafe/results/discuss/backbone.csv', 'r'))
for i, e in enumerate(f):
    if i % 3 == 1 and i / 3 < 25:
        con = e[0].split(' ')
        a, p, f = float(con[5]), float(con[13]), float(con[21])
        auroc1.append(a)
        aupr1.append(p)
        fpr1.append(f)
    elif i % 3 == 1 and i / 3 < 50:
        con = e[0].split(' ')
        a, p, f = float(con[5]), float(con[13]), float(con[21])
        auroc2.append(a)
        aupr2.append(p)
        fpr2.append(f)
    elif i % 3 == 1 and i / 3 < 75:
        con = e[0].split(' ')
        a, p, f = float(con[5]), float(con[13]), float(con[21])
        auroc3.append(a)
        aupr3.append(p)
        fpr3.append(f)

print(len(auroc1), len(auroc2), len(auroc3))

# cora label

auroc1 = np.array(auroc1).reshape(-1, 5)[:5]
aupr1 = np.array(aupr1).reshape(-1, 5)[:5]
fpr1 = np.array(fpr1).reshape(-1, 5)[:5]
print(auroc1)

# cora feature

auroc2 = np.array(auroc2).reshape(-1, 5)[:5]
aupr2 = np.array(aupr2).reshape(-1, 5)[:5]
fpr2 = np.array(fpr2).reshape(-1, 5)[:5]
print(auroc2)

# cora structure

auroc3 = np.array(auroc3).reshape(-1, 5)[:5]
aupr3 = np.array(aupr3).reshape(-1, 5)[:5]
fpr3 = np.array(fpr3).reshape(-1, 5)[:5]
print(auroc3)

fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18,3))
sns.set(style='whitegrid')

# color=['#d53e4f','#f46d43','#fdae61','#9970ab','#c2a5cf','#abdda4']
color=['#9970ab','#f46d43','#c2a5cf','#abdda4','#66c2a5']
x_name = ['MLP', 'GCN', 'GAT', 'JKNet', 'MixHop']
x_name = x_name[:5]

width=0.15
xticks = np.arange(5)
ax[0].bar(x=x_name, height=auroc1[:, 0], label='MSP', alpha=1.0, width=width, color=color[1])
ax[0].bar(x=xticks+width, height=auroc1[:, 1], label='Energy', alpha=1.0, width=width, color=color[2])
ax[0].bar(x=xticks+width*2, height=auroc1[:, 2], label='Energy FT', alpha=1.0, width=width, color=color[0])
ax[0].bar(x=xticks+width*3, height=auroc1[:, 3], label='GNNSafe', alpha=1.0, width=width, color=color[3])
ax[0].bar(x=xticks+width*4, height=auroc1[:, 4], label='GNNSafe++', alpha=1.0, width=width, color=color[4])
ax[0].set_xticks(xticks + 0.3)
ax[0].set_ylabel('AUROC', size=15)
ax[0].set_ylim([40, 100])
ax[0].set_title('Structure', size=15)
ax[0].tick_params(labelsize=15)

ax[1].bar(x=x_name, height=auroc2[:, 0], alpha=1.0, width=width, color=color[1])
ax[1].bar(x=xticks+width, height=auroc2[:, 1], alpha=1.0, width=width, color=color[2])
ax[1].bar(x=xticks+width*2, height=auroc2[:, 2], alpha=1.0, width=width, color=color[0])
ax[1].bar(x=xticks+width*3, height=auroc2[:, 3], alpha=1.0, width=width, color=color[3])
ax[1].bar(x=xticks+width*4, height=auroc2[:, 4], alpha=1.0, width=width, color=color[4])
ax[1].set_xticks(xticks + 0.3)
ax[1].set_ylabel('AUROC', size=15)
ax[1].set_ylim([40, 100])
ax[1].set_title('Feature', size=15)
ax[1].tick_params(labelsize=15)

ax[2].bar(x=x_name, height=auroc3[:, 0], alpha=1.0, width=width, color=color[1])
ax[2].bar(x=xticks+width, height=auroc3[:, 1], alpha=1.0, width=width, color=color[2])
ax[2].bar(x=xticks+width*2, height=auroc3[:, 2], alpha=1.0, width=width, color=color[0])
ax[2].bar(x=xticks+width*3, height=auroc3[:, 3], alpha=1.0, width=width, color=color[3])
ax[2].bar(x=xticks+width*4, height=auroc3[:, 4], alpha=1.0, width=width, color=color[4])
ax[2].set_xticks(xticks + 0.3)
ax[2].set_ylabel('AUROC', size=15)
ax[2].set_ylim([40, 100])
ax[2].set_title('Label', size=15)
ax[2].tick_params(labelsize=15)

fig.legend(loc='lower center', bbox_to_anchor=(0.5,0.95), fancybox=True, fontsize=15, ncol=5)
plt.tight_layout()
plt.show()
# fig.savefig('./figs/backbone.pdf', bbox_inches='tight')
```
